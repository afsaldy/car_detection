```python
from ultralytics import YOLO
import cv2
import cvzone
import math
import time
import numpy as np
from sort import Sort


# -----------------------------
# Utilities
# -----------------------------
def load_mask(mask_path: str):
    """Load mask as 1-channel grayscale (recommended for ROI masking)."""
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    if mask is None:
        raise FileNotFoundError(f"Mask tidak ditemukan: {mask_path}")
    return mask


def apply_mask(img, mask_gray):
    """Resize mask to match img and apply ROI mask."""
    mask_resized = cv2.resize(mask_gray, (img.shape[1], img.shape[0]))
    img_region = cv2.bitwise_and(img, img, mask=mask_resized)
    return img_region


def yolo_detections_to_sort(results, classNames, allowed_classes, conf_thres=0.3):
    """
    Convert YOLO results into SORT detections array:
    [[x1, y1, x2, y2, conf], ...]
    """
    detections = np.empty((0, 5), dtype=float)

    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls = int(box.cls[0])
            currentClass = classNames[cls]

            if (currentClass in allowed_classes) and (conf > conf_thres):
                currentArray = np.array([x1, y1, x2, y2, conf], dtype=float)
                detections = np.vstack((detections, currentArray))

    return detections


def draw_counting_line(img, line, color=(0, 0, 255), thickness=5):
    x1, y1, x2, y2 = line
    cv2.line(img, (x1, y1), (x2, y2), color, thickness)


def update_counter_and_draw(img, tracks, line, counter_ids, line_tolerance=20):
    """
    Draw tracked boxes + ids + center points.
    Count object if its center crosses the counting line area.
    """
    xL1, yL1, xL2, yL2 = line

    for t in tracks:
        x1, y1, x2, y2, track_id = t
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        track_id = int(track_id)

        w, h = x2 - x1, y2 - y1
        cx, cy = x1 + w // 2, y1 + h // 2

        # Draw bbox + id + center
        cvzone.cornerRect(img, (x1, y1, w, h), l=15, t=1, rt=5, colorR=(255, 0, 0))
        cvzone.putTextRect(img, f"{track_id}", (max(0, x1), max(35, y1)),
                           scale=2, thickness=3, offset=1)
        cv2.circle(img, (cx, cy), 5, (255, 0, 255), cv2.FILLED)

        # Counting condition (line is horizontal here)
        # If center is within x-range and close to y of line
        if xL1 < cx < xL2 and (yL1 - line_tolerance) < cy < (yL2 + line_tolerance):
            if track_id not in counter_ids:
                counter_ids.add(track_id)
                draw_counting_line(img, line, color=(0, 255, 0), thickness=5)


def draw_fps(img, prev_time):
    """Return updated prev_time and draw FPS on frame."""
    now = time.time()
    fps = 1 / max(now - prev_time, 1e-6)
    cvzone.putTextRect(img, f"FPS: {int(fps)}", (20, 50), scale=1, thickness=2)
    return now


# -----------------------------
# Main
# -----------------------------
def main():
    video_path = r"D:\Belajar FR\Youtube_2\video\cars.mp4"
    mask_path = r"mask.png"
    model_path = "yolo11s.pt"

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise FileNotFoundError(f"Video tidak bisa dibuka: {video_path}")

    model = YOLO(model_path)

    classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
                  "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
                  "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
                  "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
                  "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
                  "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
                  "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
                  "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
                  "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
                  "teddy bear", "hair drier", "toothbrush"]

    allowed_classes = {"car", "motorbike", "bus", "truck"}
    conf_thres = 0.3

    mask_gray = load_mask(mask_path)
    tracker = Sort()

    # line format: [x1, y1, x2, y2]
    line = [185, 400, 750, 400]

    counter_ids = set()
    prev_time = time.time()

    while True:
        success, img = cap.read()
        if not success or img is None:
            print("Video selesai / frame tidak terbaca.")
            break

        # apply ROI mask
        imgRegion = apply_mask(img, mask_gray)

        # YOLO detection on region
        results = model(imgRegion, stream=True)

        # to SORT format
        detections = yolo_detections_to_sort(
            results, classNames, allowed_classes, conf_thres=conf_thres
        )

        # update tracker
        tracks = tracker.update(detections)

        # draw default line (red)
        draw_counting_line(img, line, color=(0, 0, 255), thickness=5)

        # draw tracks + counting
        update_counter_and_draw(img, tracks, line, counter_ids, line_tolerance=20)

        # draw counter
        cvzone.putTextRect(img, f"Counter: {len(counter_ids)}", (50, 100),
                           scale=2, thickness=1, offset=20)

        # fps
        prev_time = draw_fps(img, prev_time)

        cv2.imshow("image", img)
        # cv2.imshow("imageRegion", imgRegion)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()


if __name__ == "__main__":
    main()
```
